```r
# Set graphical parameters
par(bg = "white", 
    cex.axis = 1.5, 
    cex.lab = 1.7, 
    cex.main = 2, 
    mfrow = c(1, 3),  # Arrange plots in 1 row and 3 columns
    mar = c(5, 5, 4, 2) + 0.1,  # Adjust margins (bottom, left, top, right)
    oma = c(0, 0, 0, 2),  # Add space on the right side
    mgp = c(3, 1, 0))  # Adjust axis label spacing

# Define number of breaks and axis limits
num_breaks <- seq(1, 4, by = 0.5)  # Adjust "by" value to control number of bars
x_limits <- c(1, 4)
y_limits <- c(0, 40)

# Plot for stage 1
hist(stage1$agg_MOS, 
     col = "lightblue",
     border = "black", 
     main = "Stage 1 MOS Distribution",  # Title
     xlab = "(a) Stage 1 MOS",  # X-axis label
     ylab = "Frequency",  # Y-axis label
     breaks = num_breaks,  # Set consistent breaks
     xlim = x_limits,  # Set x-axis limits
     ylim = y_limits)  # Consistent y-axis limits

# Plot for stage 2
hist(stage2$agg_MOS, 
     col = "lightblue",
     border = "black", 
     main = "Stage 2 MOS Distribution", 
     xlab = "(b) Stage 2 MOS", 
     ylab = "Frequency", 
     breaks = num_breaks, 
     xlim = x_limits, 
     ylim = y_limits)

# Plot for stage 3
hist(stage3$agg_MOS, 
     col = "lightblue",
     border = "black", 
     main = "Stage 3 MOS Distribution", 
     xlab = "(c) Stage 3 MOS", 
     ylab = "Frequency", 
     breaks = num_breaks, 
     xlim = x_limits, 
     ylim = y_limits)
```
