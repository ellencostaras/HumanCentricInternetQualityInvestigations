```python
def run_subjective_ellen(conditions):

    #click chrome
    pyautogui.click(202, 860)
    
    #click into student chrome profile
    pyautogui.click(725, 511)

    #click into search bar, type webRTC internals site and press enter
    pyautogui.click(386, 89)
    pyautogui.write('chrome://webrtc-internals/')
    pyautogui.press('enter')

    #click new tab
    pyautogui.click(344, 41)

    #type google meet link and press enter 
    pyautogui.write('meet.google.com/ntx-rrgc-rrb')
    pyautogui.press('enter')

    #open terminal 
    pyautogui.click(1116, 877)
    
    up = 50000
    down = 100000
    rtt = 50

    #type throttle conditions and press enter
    pyautogui.write(f'throttle --up {up} --down {down} --rtt {rtt}')
    pyautogui.press('enter')

    #type password and press enter 
    pyautogui.write('050902Ec')
    pyautogui.press('enter')

    #wait a beat for throttle to kick in 
    time.sleep(10)

    #click chrome 
    pyautogui.click(252, 909)

    #join call 
    pyautogui.click(1115, 515)
    pyautogui.click(1115, 515)

    #wait 30 seconds
    time.sleep(30) 

    #hang up
    pyautogui.click(926, 793)
    pyautogui.click(817, 523)

    #wait to ensure hang up is correct 
    time.sleep(5)

    #exit meet tab
    pyautogui.click(546, 44)

    #reopen terminal 
    pyautogui.click(1116, 877)

    #type stop throttle and press enter
    pyautogui.write('throttle stop')
    pyautogui.press('enter')

    #quit terminal
    pyautogui.click(65, 8)
    pyautogui.click(125, 240)

    #click chrome
    pyautogui.click(252, 909)

    #download WebRTC dump 
    pyautogui.click(9, 135)
    pyautogui.click(40, 155)

    #quit chrome
    pyautogui.click(65, 8)
    pyautogui.click(88, 309)

    #rename the stat dump file and move into storage
    treatment_number = conditions['treatment_number']
    source_path = '/Users/ellencostaras/Downloads/webrtc_internals_dump.txt'
    if os.path.isfile(source_path):
        destination_path = f'/Users/ellencostaras/Desktop/stat_dumps/treatment{treatment_number}_ellen.txt' 
        shutil.move(source_path, destination_path) 
    else:
        print(f'Warning: Stat dump for treatment {treatment_number} did not download correctly.')


def run_subjective_aadya(conditions):
    
    #click chrome
    pyautogui.click(252, 909)

    #click profile
    pyautogui.click(735, 525) 

    #click into search bar, type webRTC internals site and press enter
    pyautogui.click(386, 89)
    pyautogui.write('chrome://webrtc-internals/')
    pyautogui.press('enter')

    #click new tab
    pyautogui.click(344, 41)

    #type google meet link and press enter 
    pyautogui.write('meet.google.com/ntx-rrgc-rrb')
    pyautogui.press('enter')

    #open terminal 
    pyautogui.click(1005, 924)
    
    up = conditions['upload']
    down = conditions['download']
    rtt = conditions['latency']

    #type throttle conditions and press enter
    pyautogui.write(f'throttle --up {up} --down {down} --rtt {rtt}')
    pyautogui.press('enter')

    #type password and press enter 
    pyautogui.write('bahhumbug28')
    pyautogui.press('enter')

    #wait a beat for throttle to kick in 
    time.sleep(10)

    #click chrome 
    pyautogui.click(252, 909)

    #click join call 
    pyautogui.click(1115, 547) 

    #wait 30 seconds
    time.sleep(30) 

    #wait a beat for the meeting to be completely over 
    time.sleep(8)

    #exit meet tab
    pyautogui.click(546, 57)

    #reopen terminal 
    pyautogui.click(1005, 924)

    #type throttle stop and press enter
    pyautogui.write('throttle stop')
    pyautogui.press('enter')

    #quit terminal 
    pyautogui.click(65, 8)
    pyautogui.click(125, 240)

    #click chrome
    pyautogui.click(252, 909)

    #download WebRTC dump 
    pyautogui.click(39, 175)
    pyautogui.click(67, 203)

    #quit chrome
    pyautogui.click(77, 16)
    pyautogui.click(90, 316)

    #rename the stat dump file and move into storage
    treatment_number = conditions['treatment_number']
    source_path = '/Users/User/Downloads/webrtc_internals_dump.txt'
    if os.path.isfile(source_path):
        destination_path = f'/Users/User/Desktop/stat_dumps/treatment{treatment_number}_aadya.txt' 
        shutil.move(source_path, destination_path) 
    else:
        print(f'Warning: Stat dump for treatment {treatment_number} did not download correctly.')


def run_subjective_experiment(user:str, first:int, last:int):

    pyautogui.FAILSAFE = False
    pyautogui.PAUSE = 2 #adds a 2 second delay between each instruction
    treatments_file_path = "/Users/User/Documents/GitHub/DesktopDownthrottlingExperiment/pre_experiment/subjective_combos_shuffled_aadya.csv"
    
    treatments = []
    with open(treatments_file_path, 'r') as file:
        csvreader = csv.reader(file)
        header = next(csvreader)
        for row in csvreader:
            treatments.append(row)

    for treatment in treatments[first:last]:

        now = datetime.datetime.now()
        if now.minute >= 57:
            next_minute = (now.minute + 1) % 60
            next_hour = (now.hour + (now.minute + 1) // 60) % 24
        else:
            next_minute = now.minute + 1
            next_hour = now.hour
        next_minute_mark = now.replace(hour=next_hour, minute=next_minute, second=0, microsecond=0)
        time_to_wait = (next_minute_mark - now).total_seconds()
        print(f"Waiting until {next_minute_mark.time()} to begin the next treatment.")
        time.sleep(time_to_wait)

        conditions = {
            "treatment_number": int(treatment[0]),
            "upload": int(float(treatment[1])),
            "download": int(treatment[2]), 
            "latency": int(treatment[3])}

        if user == "Ellen":
            try:
                run_subjective_ellen(conditions)
                print("Treatment", conditions['treatment_number'], "completed.")
            except Exception as e:
                error_type = type(e).__name__
                print(f'Treatment {conditions["treatment_number"]} failed.\nError: {error_type} - {str(e)}')
                print('Beginning next treatment.')
        
        elif user == "Aadya":
            try:
                run_subjective_aadya(conditions)
                print("Treatment", conditions['treatment_number'], "completed.")
            except Exception as e:
                error_type = type(e).__name__
                print(f'Treatment {conditions["treatment_number"]} failed.\nError: {error_type} - {str(e)}')
                print('Beginning next treatment.')

    return " --------------- Subjective monitoring complete! --------------- "
```

```python
user = "Ellen"
first = 0
last = 12
run_subjective_experiment(user, first, last)
```
